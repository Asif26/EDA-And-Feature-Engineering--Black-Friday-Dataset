## Balck Friday Dataset EDA And Feature Engineering
### cleaning and perparing the Data for model traning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
%matplotlib inline
```

```python
# importing Datasets
df_train=pd.read_csv("./train.csv")
df_test=pd.read_csv("./test.csv")
```

```python
df_train.head()
```

```python
## merge both train and test daata
df = df_train.append(df_test)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('User_ID',axis=1, inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'],drop_first=1)
```

```python
## Handling categorical feature
df["Gender"]=df['Gender'].map({'F':0,'M':1})
```

```python
df.head()
```

```python
# Handling categorical feature Age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
## 2nd Technqiue
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])

df['Age'].unique()
```

```python
## fixing categorical City_catgory
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
df = pd.concat([df,df_city],axis=1)
```

```python
df.head()
```

```python
# drop City_Category 
df.drop('City_Category',axis=1, inplace=True)
```

```python
df
```

```python
# Mising Values
df.isnull().sum()
```

```python
# Product_Category_2 replace missing values
df['Product_Category_2']=df['Product_Category_2'].fillna(df["Product_Category_2"].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
# Product_Category_3 replace missing values
df['Product_Category_3']= df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace("+","")
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
# convert object into integers
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```

```python
df.info()
```
